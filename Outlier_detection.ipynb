```python
# loading the diabetes dataset from sklearn library
from sklearn.datasets import load_diabetes
```

```python
dataset= load_diabetes()
```

```python
dataset.feature_names
```

```python
import pandas as pd
import numpy as np
```

```python
df= pd.DataFrame(data=np.c_[dataset.data,dataset.target],columns=dataset.feature_names + ['target'])
df.head()
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().any() #no missing value present
```
